```python
columns = ['Code du département', 'Code de la commune', 'Inscrits', '% Abs/Ins', '% Vot/Ins',
           '% Blancs/Ins', '% Blancs/Vot', '% Nuls/Ins', '% Nuls/Vot', '% Exp/Ins', '% Exp/Vot']
duplicatas = ['Sexe', 'Nom', 'Voix', '% Voix/Ins', '% Voix/Exp']
elections = pandas.DataFrame(columns=columns+duplicatas)
dfs = []
for i in range(1, 11):
    # f = string formatting
    lst = columns + [f'{j}.{i}' for j in duplicatas]
    elections = pandas.DataFrame(columns=columns+duplicatas)
    elections[columns + duplicatas] = df2017_1[lst]
    dfs.append(elections)
elections = pandas.concat([elections] + dfs)
df2017_1 = elections
df2017_1['Tour'] = 0
print(df2017_1.shape)
df2017_1.head()
```

```python
sanctions.isnull().sum()/len(sanctions)
```

```python
sanctions = sanctions.drop(['conseil_discipline','type_sanction'], axis=1)
sanctions['sexe'] = sanctions['sexe'].fillna(random.choice(['F','H']))
sanctions['sexe'] = sanctions['sexe'].replace('f', 'F')
sanctions['sexe'] = sanctions['sexe'].replace('h', 'H')
sanctions['nombre'] = sanctions['nombre'].fillna(sanctions['nombre'].mean())
```

```python
naissances = naissances[['ANAIS', 'ACCOUCHR', 'AGEXACTM', 'AGEXACTP',
       'AMAR', 'ARECC', 'ARECM', 'ARECP', 'DEPDOM', 'DEPNAIS', 'DURECEVP',
       'INDLNM', 'INDLNP', 'INDNATM', 'INDNATP', 'JRECC', 'JRECM', 'JRECP',
       'MNAIS', 'MRECC', 'MRECM', 'MRECP', 'NBENF', 'ORIGINOM', 'SEXE',
       'SITUATMR', 'SITUATPR']]
naissances
```

```python
pourcentage_nulles = naissances.isnull().mean()
colonnes_a_supprimer = pourcentage_nulles[pourcentage_nulles > 0.4].index
naissances = naissances.drop(colonnes_a_supprimer, axis=1)
naissances
```

```python
naissances['DURECEVP'] = naissances['DURECEVP'].fillna(lambda x: random.choice(naissances['DURECEVP']))
naissances['ACCOUCHR'] = naissances['ACCOUCHR'].fillna(lambda x: random.choice(naissances['ACCOUCHR']))
naissances['ORIGINOM'] = naissances['ORIGINOM'].fillna(lambda x: random.choice(naissances['ORIGINOM']))
naissances['SITUATMR'] = naissances['SITUATMR'].fillna(lambda x: random.choice(naissances['SITUATMR']))
naissances['SITUATPR'] = naissances['SITUATPR'].fillna(lambda x: random.choice(naissances['SITUATPR']))
```

```python
df_chomage = pandas.read_csv("./chomage/gironde/valeurs_trimestrielles.csv", sep=";")
df_chomage.head()
```

```python
df_chomage = df_chomage.drop('Dernière mise à jour', axis=1)
df_chomage = df_chomage.drop('idBank', axis=1)
df_chomage = df_chomage.drop('Période', axis=1)
df_chomage = df_chomage.drop(index=1)
df_chomage.at[0,"Libellé"] = 33
df_chomage.rename(columns={'Libellé': 'Code du département'}, inplace=True)
df_chomage.head()
```

```python
for column in df_chomage:
    df_chomage[column] = pandas.to_numeric(df_chomage[column], errors='coerce')
df_chomage['moyenne_2021'] = df_chomage[["2021-T1","2021-T2","2021-T3","2021-T4"]].mean(axis=1)
df_chomage['moyenne_2020'] = df_chomage[["2020-T1","2020-T2","2020-T3","2020-T4"]].mean(axis=1)
df_chomage['moyenne_2019'] = df_chomage[["2019-T1","2019-T2","2019-T3","2019-T4"]].mean(axis=1)
df_chomage['moyenne_2018'] = df_chomage[["2018-T1","2018-T2","2018-T3","2018-T4"]].mean(axis=1)
df_chomage['moyenne_2017'] = df_chomage[["2017-T1","2017-T2","2017-T3","2017-T4"]].mean(axis=1)
columns_to_drop = df_chomage.filter(regex='T[1-4]').columns
df_chomage.drop(columns=columns_to_drop, inplace=True)
df_chomage.head()
```

```python
df_chomage.at[0, "année"] = 2017
df_chomage.at[1, "année"] = 2018
df_chomage.at[2, "année"] = 2019
df_chomage.at[3, "année"] = 2020
df_chomage.at[4, "année"] = 2021

df_chomage.at[0, "moyenne_chomage"] = df_chomage.at[0, "moyenne_2017"]
df_chomage.at[1, "moyenne_chomage"] = df_chomage.at[0, "moyenne_2018"]
df_chomage.at[2, "moyenne_chomage"] = df_chomage.at[0, "moyenne_2019"]
df_chomage.at[3, "moyenne_chomage"] = df_chomage.at[0, "moyenne_2020"]
df_chomage.at[4, "moyenne_chomage"] = df_chomage.at[0, "moyenne_2021"]

df_chomage.at[0, "Code du département"] = "33"
df_chomage.at[1, "Code du département"] = "33"
df_chomage.at[2, "Code du département"] = "33"
df_chomage.at[3, "Code du département"] = "33"
df_chomage.at[4, "Code du département"] = "33"

columns_to_drop = df_chomage.filter(regex='moyenne_[1-9]').columns
df_chomage.drop(columns=columns_to_drop, inplace=True)
df_chomage.head()


df_chomage.head()
```
