# 2017 - TOUR 1

```python
df2017_1 = pd.read_excel("presidentielles/2017/1er_tour/Presidentielle_2017_Resultats_Communes_Tour_1.xls", skiprows=3)
```

```python
print(df2017_1.shape)
df2017_1.head()
```

```python
columns = ['Code du département', 'Code de la commune', 'Inscrits', '% Abs/Ins', '% Vot/Ins',
           '% Blancs/Ins', '% Blancs/Vot', '% Nuls/Ins', '% Nuls/Vot', '% Exp/Ins', '% Exp/Vot']
duplicatas = ['Sexe', 'Nom', 'Voix', '% Voix/Ins', '% Voix/Exp']
df = pd.DataFrame(columns=columns+duplicatas)
dfs = []
for i in range(1, 11):
    # f = string formatting
    lst = columns + [f'{j}.{i}' for j in duplicatas]
    df = pd.DataFrame(columns=columns+duplicatas)
    df[columns + duplicatas] = df2017_1[lst]
    dfs.append(df)
df = pd.concat([df] + dfs)
df2017_1 = df
df2017_1['Tour'] = 0
print(df2017_1.shape)
df2017_1.head()
```

# Sanctions

```python
df_sanction = pdor.read_ods("sanction_police/sanctionspolice-1990-2021.ods")
df_sanction.head()
```

```python
df_sanction.isnull().sum()/len(df_sanction)
```

```python
df_sanction = df_sanction.drop(['conseil_discipline','type_sanction'], axis=1)
df_sanction['sexe'] = df_sanction['sexe'].fillna(random.choice(['F','H']))
df_sanction['sexe'] = df_sanction['sexe'].replace('f', 'F')
df_sanction['sexe'] = df_sanction['sexe'].replace('h', 'H')
df_sanction['nombre'] = df_sanction['nombre'].fillna(df_sanction['nombre'].mean())
df_sanction['motif'] = df_sanction['motif'].fillna(random.choice(df_sanction['motif']))
```

# Naissances

```python
naissance = pd.read_csv('naissances/etatcivil2019_nais2019_csv/FD_NAIS_2019.csv', delimiter=';')
naissance
```

```python
naissance = naissance[['ANAIS', 'ACCOUCHR', 'AGEXACTM', 'AGEXACTP',
       'AMAR', 'ARECC', 'ARECM', 'ARECP', 'DEPDOM', 'DEPNAIS', 'DURECEVP',
       'INDLNM', 'INDLNP', 'INDNATM', 'INDNATP', 'JRECC', 'JRECM', 'JRECP',
       'MNAIS', 'MRECC', 'MRECM', 'MRECP', 'NBENF', 'ORIGINOM', 'SEXE',
       'SITUATMR', 'SITUATPR']]
naissance
```

```python
pourcentage_nulles = naissance.isnull().mean()
colonnes_a_supprimer = pourcentage_nulles[pourcentage_nulles > 0.4].index
naissance.drop(colonnes_a_supprimer, axis=1, inplace=True)
naissance
```

```python
naissance['DURECEVP'] = naissance['DURECEVP'].fillna(random.choice(naissance['DURECEVP']))
naissance['ACCOUCHR'] = naissance['ACCOUCHR'].fillna(random.choice(naissance['ACCOUCHR']))
naissance['ORIGINOM'] = naissance['ORIGINOM'].fillna(random.choice(naissance['ORIGINOM']))
naissance['SITUATMR'] = naissance['SITUATMR'].fillna(random.choice(naissance['SITUATMR']))
naissance['SITUATPR'] = naissance['SITUATPR'].fillna(random.choice(naissance['SITUATPR']))
naissance
```
