```python
import pandas as pd
import numpy  as np
```

# 2017 - TOUR 1

```python
df2017_1 = pd.read_excel("presidentielles/2017/1er_tour/Presidentielle_2017_Resultats_Communes_Tour_1.xls", skiprows=3)
```

```python
print(df2017_1.shape)
df2017_1.head()
```

```python
columns = ['Code du département', 'Code de la commune', 'Inscrits', '% Abs/Ins', '% Vot/Ins',
           '% Blancs/Ins', '% Blancs/Vot', '% Nuls/Ins', '% Nuls/Vot', '% Exp/Ins', '% Exp/Vot']
duplicatas = ['Sexe', 'Nom', 'Voix', '% Voix/Ins', '% Voix/Exp']
df = pd.DataFrame(columns=columns+duplicatas)
dfs = []
for i in range(1, 11):
    lst = columns + [f'{j}.{i}' for j in duplicatas]
    df = pd.DataFrame(columns=columns+duplicatas)
    df[columns + duplicatas] = df2017_1[lst]
    dfs.append(df)
df = pd.concat([df] + dfs)
df2017_1 = df
df2017_1['Tour'] = 0
print(df2017_1.shape)
df2017_1.head()
```

# 2022 - TOUR 1

```python
df2022_1 = pd.read_excel("presidentielles/2022/1er_tour/resultats-par-niveau-subcom-t1-france-entiere.xlsx")
```

```python
print(df2022_1.shape)
df2022_1.head()
```

```python
rename_dict = {}
duplicatas = ['N°Panneau', 'Sexe', 'Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp']
for i, col_name in enumerate(df2022_1.columns[df2022_1.columns.str.contains('^Unnamed')].values):
    index = i % len(duplicatas)
    new_col_name = duplicatas[index] + f'.{((i + len(duplicatas)) // len(duplicatas))}'
    rename_dict[col_name] = new_col_name

df2022_1.rename(columns=rename_dict, inplace=True)
df2022_1.head()
```

```python
duplicatas = ['Sexe', 'Nom', 'Voix', '% Voix/Ins', '% Voix/Exp']
df = pd.DataFrame(columns=columns+duplicatas)
dfs = []
for i in range(1, 12):
    lst = columns + [f'{j}.{i}' for j in duplicatas]
    df = pd.DataFrame(columns=columns+duplicatas)
    df[columns + duplicatas] = df2022_1[lst]
    dfs.append(df)
df = pd.concat([df] + dfs)
df2022_1 = df
df2022_1['Tour'] = 0
print(df2022_1.shape)
df2022_1.head()
```
